```python
import pandas as pd
import requests
import mysql.connector
```

# Read data from database (Sql)

```python
#read data from a data base and convert it to a pandas data frame
conn=mysql.connector.connect(host='localhost',user='root',password='',database='customer_database')
customer=pd.read_sql_query('SELECT * FROM customers',conn)
orders=pd.read_sql_query('SELECT * FROM orders',conn)
products=pd.read_sql_query('SELECT * FROM products',conn)
```

```python
customer
```

```python
with pd.ExcelWriter('market.xlsx') as writer:
    customer.to_excel(writer,sheet_name='customer_sheet')
    orders.to_excel(writer,sheet_name='order_sheet')
    products.to_excel(writer,sheet_name='products')
```
