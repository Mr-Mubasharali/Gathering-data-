```python
# Apply 0.5 timer because if we hit a lot of requests in miliseconds the website of api block they assume that is a attach:This is a rule that says you can only make a certain number of requests per second or per minute.
import time
for i in range(1,200):
    
    url='https://wft-geo-db.p.rapidapi.com'
    page_n= url+next_url
    response = requests.get(page_n, headers=headers)
    # check response status on every itteration if there is any error they will response.
    if response.status_code !=200:
        print(f"Ther is some issue on the {i} itteration this error code is {response.status_code}")
        break
    res_json=response.json()
    links=res_json['links']
    for item in links:
        if item['rel']=='next':
            next_url=item['href']
    data.extend(res_json['data'])
    time.sleep(0.5)
    
    
```

```python
df=pd.DataFrame(data)
```
