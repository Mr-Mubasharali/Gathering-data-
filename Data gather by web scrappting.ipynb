```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import lxml
import numpy as np
import random
```

# Gather data by using Web scrapping

```python
url="https://priceoye.pk/mobiles?page=1"
#want to HTML content use text
Session=requests.Session()
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.5",
    "Referer": "https://www.google.com/",
    "DNT": "1", # Do Not Track request header
    "Connection": "keep-alive",
    "Upgrade-Insecure-Requests": "1"
}
responce=Session.get(url,headers=headers).text
```

```python
soup=BeautifulSoup(responce,"lxml")
```

```python
all_div=soup.find_all('div',class_="productBox b-productBox")
```

```python
img=[]
phone_name=[]
rating=[]
reviews=[]
discount_price=[]
original_price=[]
for phone in all_div:
   try:
      img.append(phone.find('img')['src'])
   except:
      img.append(np.nan)
   try:
      phone_name.append(phone.find('div',class_='text-box').text)
   except:
      phone_name.append(np.nan)
   try:
      rating.append(float(phone.find('span',class_="h6 bold").get_text(strip=True)))
   except:
      rating.append(np.nan)
   try:
      reviews.append(int(phone.find('span',class_="rating-h7 bold space").text))
   except:
      reviews.append(np.nan)
   try:
      discount_price.append(phone.find('div',class_="pricing-setting").find_all('span')[0].text)
   except:
      discount_price.append(np.nan)
   try:
      original_price.append(phone.find('div',class_="pricing-setting").find_all('span')[1].text)
   except:
      original_price.append(np.nan)
       
   
   

    
    
   
```

```python
df=pd.DataFrame({'Phone_name':phone_name,'Rating':rating,'Reviews':reviews,'Phone_price':original_price,'Discounted_price':discount_price,'Phone_image':img})
```

```python
mobile_df=pd.DataFrame()


for i in range(1,21):
    # Each itration change the web page number
    url=f"https://priceoye.pk/mobiles?page={i}"
    responce=Session.get(url,headers=headers).text
    soup=BeautifulSoup(responce,"lxml")
    all_div=soup.find_all('div',class_="productBox b-productBox")
    img=[]
    phone_name=[]
    rating=[]
    reviews=[]
    discount_price=[]
    original_price=[]
    for phone in all_div:
        try:
            img.append(phone.find('img')['src'])
        except:
            img.append(np.nan)
        try:
            phone_name.append(phone.find('div',class_='text-box').text)
        except:
            phone_name.append(np.nan)
        try:
            rating.append(float(phone.find('span',class_="h6 bold").get_text(strip=True)))
        except:
            rating.append(np.nan)
        try:
            reviews.append(int(phone.find('span',class_="rating-h7 bold space").text))
        except:
            reviews.append(np.nan)
        try:
            discount_price.append(phone.find('div',class_="pricing-setting").find_all('span')[0].text)
        except:
            discount_price.append(np.nan)
        try:
            original_price.append(phone.find('div',class_="pricing-setting").find_all('span')[1].text)
        except:
            original_price.append(np.nan)
    df=pd.DataFrame({'Phone_name':phone_name,'Rating':rating,'Reviews':reviews,'Phone_price':original_price,'Discounted_price':discount_price,'Phone_image':img})
    mobile_df=pd.concat([mobile_df,df],ignore_index=True)
   # Each itration clear all the lists
    img.clear()
    phone_name.clear()
    rating.clear()
    reviews.clear()
    discount_price.clear()
    original_price.clear()
    # Random delay for 1 to 3 sec for safe scrapping
    time.sleep(random.uniform(1,3))
    
```
